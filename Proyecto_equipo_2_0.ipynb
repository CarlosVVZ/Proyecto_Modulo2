```python
#LIMPIEZA DEL DF

# Creating a copy of the original dataframe- df
df1 = file.copy()
df1.drop('S.NO',axis=1,inplace=True)
df1.info()
```

```python
#FORMATO DEL DF

# Convert string to datetime64
df1['Year'] = df1['Year'].apply(pd.to_datetime,format='%Y')

#Set Date column as the index column.
df1['year'] = pd.DatetimeIndex(df1['Year']).year  
df1.set_index('year', inplace=True)
df1.drop('Year',axis=1,inplace=True)

# Converting the sport column to uppercase
df1['Sport'] = df1['Sport'].str.upper()
df1.head()
```

```python
df_rnk = df1.copy() #Copiamos el DF limpio
df_rnk["Previous Year Rank"] = df_rnk["Previous Year Rank"].astype(str).str.extract('(\d+)', expand=False) #Limpiar columna de ranking año pasado (quitar todo menos números)
df_rnk = df_rnk[(df_rnk['Current Rank'].notna()) & (df_rnk['Previous Year Rank'].notna())] # Filtrar filas, sólo podemos contrastar el ranking si ambas columnas tienen valor asignado
df_rnk["Previous Year Rank"] = df_rnk["Previous Year Rank"].astype(int) #Convertir ranking año pasado a entero
df_rnk["Current Rank"] = df_rnk["Current Rank"].astype(int) #Convertir ranking año actual a entero
data_10_20 = df_rnk[(df_rnk.index >= 2010) & (df_rnk.index <= 2020) & ( df_rnk["Previous Year Rank"] - df_rnk["Current Rank"]  >= 2)] 
#Filtrar filas, si el año es mayor o igual a 2010 y menor o igual a 2020, también si la diferencia entre el ranking anterior y el actual es mayor o igual a 2
data_10_20 = data_10_20.sort_values(by='year',ascending=True)

z = data_10_20[['Name','Sport',"Previous Year Rank","Current Rank"]]
z
```

```python
#Apariciones por ganancia
apariciones = file.groupby(['Name']).Name.count()
aparciones = pd.DataFrame(apariciones)
dinero = file.groupby(['Name']).earnings.sum()
dinero = pd.DataFrame(dinero)
df_dinero=pd.concat([apariciones,dinero],join ='inner', axis=1)
df_dinero= df_dinero.rename(columns= {'Name':'Apariciones'})
df_dinero= df_dinero.sort_values(['earnings','Apariciones'],ascending=False).sort_values('Apariciones')
df_dinero
```

```python
#Atleta que gana mas con menos apariciones
atleta = df_dinero.loc[df_dinero['Apariciones']==1]
atleta= atleta.sort_values('earnings')
print('El atleta que mas gana con menos apariciones es:\n\n', atleta[-1:])
```

```python
#Lista de numero de jugadores por pais y deporte
Lista_paises=[]
for i in file['Nationality']:
    if i in Lista_paises:
        pass
    else:
        Lista_paises.append(i)
for i in Lista_paises:
    Pais_ = file.loc[file['Nationality']== i]
    df_dep_pais=Pais_.groupby(['Sport']).Sport.count()
    print(i)
    print(df_dep_pais,'\n\n')
```

```python
#Lista de minima y maxima ganancia por pais
lista_menor=[]
lista_mayor=[]
for i in Lista_paises:
    Pais_ = file.loc[file['Nationality']== i].sort_values(by = 'earnings')
    Pais_= Pais_.drop(columns= ['S.NO','Current_Rank','Previous_Year_Rank','Sport'])
    #print(i,'\n')
    lista_menor.append(Pais_.head(1))
    lista_mayor.append(Pais_.tail(1))
    #print('Menor ganancia\n',Pais_[:1],'\n\n')
    #print('Mayor ganancia\n',Pais_[-1:],'\n\n')
pd.DataFrame({'A':lista_menor,'B':lista_mayor})
a=pd.concat(lista_menor) 
b=pd.concat(lista_mayor)
c=pd.concat([a,b]).sort_values(by=['Nationality','earnings'])
c
```

```python
#Lista de menor y mayor ganancia por deporte
Lista_deportes=[]
for i in file['Sport']:
    if i in Lista_deportes:
        pass
    else:
        Lista_deportes.append(i)
for i in Lista_deportes:
    dep_ = file.loc[file['Sport']== i].sort_values(by = 'earnings')
    dep_= dep_.drop(columns= ['S.NO','Nationality','Current_Rank','Previous_Year_Rank','Nationality'])
    print(i,'\n')
    print('Menor ganancia\n',dep_[:1],'\n\n')
    print('Mayor ganancia\n',dep_[-1:],'\n\n')
```

```python
#Jugador con mayor ganancia por decada
Decadas= ['90','2000','2010','2020']
file_=file.copy()
file_= file_.drop(columns= ['S.NO','Nationality','Current_Rank','Previous_Year_Rank','Nationality'])
```

```python
file_['Year']=pd.cut(file['Year'],bins=[1990,2000,2010,2019,2020], labels= Decadas)
file_= file_.rename(columns= {'Year':'Decada'})
file_
```

```python
#Jugador con mayor ganancia por decada
for i in Decadas:
    dec_ = file_.loc[file_['Decada']== i].sort_values(by = 'earnings')
    print('Decada',i)
    print(dec_[-1:],'\n\n')
```

```python
#Ganancia total por cada deporte por cada año
Lista_anhos = []
for i in file['Year']:
    if i in Lista_anhos:
        pass
    else:
        Lista_anhos.append(i)
for i in Lista_anhos:
    anho = file.loc[file['Year']== i].sort_values(by = 'earnings')
    print(i,'\n')
    tabla=anho.groupby(['Sport']).earnings.sum()
    print(tabla,'\n\n')
```
